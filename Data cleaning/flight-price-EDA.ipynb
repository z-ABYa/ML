```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df[['date', 'month', 'year']] = df[['date', 'month', 'year']].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['arrival_hour'] = df['arrival_hour'].astype(int) 
df['arrival_min'] = df['arrival_min'].astype(int) # can do both in same line like before
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
df[['Dep_hour', 'Dep_min']] = df[['Dep_hour', 'Dep_min']].astype(int)
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
df.info()
```

```python
df['Total_Stops'].head(3)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isna().sum() # o/p = 1
df[df['Total_Stops'].isna()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4, np.nan:1})
```

```python
df.head(3)
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head(3)
```

```python
df_hour = df['Duration'].str.split(' ').str[0]
df_min = df['Duration'].str.split(' ').str[1]

df['duration_hour'] = df_hour.str.split('h').str[0]
df['duration_min'] = df_min.str.split('m').str[0].fillna(0)
```

```python
df.drop('Duration', axis=1, inplace=True)
```
